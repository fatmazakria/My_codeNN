```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
dataframe=pd.read_csv("iris_train_features.csv",header=None)
dataset=dataframe.values
X=dataset[:,0:4].astype(float)
```

```python
dataframe=pd.read_csv("iris_train_label.csv", header=None)
dataset=dataframe.values
Y=dataset[:,0:3].astype(int)
```

```python
input_layer=1
#hidden_layer1=1
#hiddden_layer2=4
output_layer=3
```

```python
model=tf.keras.Sequential();
model.add(tf.keras.layers.Dense(input_layer,input_shape=(4,),name='input_layer',activation=tf.nn.relu,kernel_regularizer='l1'))
model.add(tf.keras.layers.Dense(output_layer,name='out_layer',activation='softmax',kernel_regularizer='l1'))
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
#writer=tf.summary.FileWriter('./graphs', tf.get_default_graph()))
model.fit(X,Y,epochs=1000,batch_size=105)
model.get_weights()

```

```python
W=model.get_weights()
w1=tf.constant(W[0])
b1=tf.constant(W[1])
w2=tf.constant(W[2])
b2=tf.constant(W[3])
x=tf.placeholder(tf.float32)
h=tf.maximum(0.0,tf.add(tf.matmul(x,w1),b1))
out=tf.nn.softmax(tf.add(tf.matmul(h,w2),b2))
error=[]

with tf.Session() as sess:
    for i in  range(105): 
        values={x:X[i,:].reshape(1,4)}
        print(sess.run(out,feed_dict=values),Y[i,])



```
